```python
import cv2
import numpy as np

# Load the image
image = cv2.imread('boxes1.jpg')

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply GaussianBlur to reduce noise and help contour detection
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Use adaptive thresholding instead of Canny edge detection
_, thresh = cv2.threshold(blurred, 200, 255, cv2.THRESH_BINARY)

# Find contours in the thresholded image
contours, _ = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

# Print the number of contours found
print(f'Number of contours found: {len(contours)}')

# Iterate through contours and print their area
for i, cnt in enumerate(contours):
    area = cv2.contourArea(cnt)
    print(f'Contour {i + 1} area: {area}')

# Optionally, you can draw the contours on the original image
cv2.drawContours(image, contours, -1, (0, 255, 0), 2)

# Display the original image with contours
cv2.imshow('Contours', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
