```python
def draw_rush(x, y, tl, tr, bl, br, h, v):
    if x <= 0 or y <= 0:
        return ""

    lines = []
    for row in range(y):
        line = ""
        for col in range(x):
            if row == 0 and col == 0:
                line += tl
            elif row == 0 and col == x - 1:
                line += tr
            elif row == y - 1 and col == 0:
                line += bl
            elif row == y - 1 and col == x - 1:
                line += br
            elif row == 0 or row == y - 1:
                line += h
            elif col == 0 or col == x - 1:
                line += v
            else:
                line += " "
        lines.append(line)
    return "\n".join(lines)
```

```python
def rush00(x, y):
    return draw_rush(x, y, "o", "o", "o", "o", "-", "|")
```

```python
def rush01(x, y):
    return draw_rush(x, y, "/", "\\", "\\", "/", "*", "*")
```

```python
def rush02(x, y):
    return draw_rush(x, y, "A", "A", "C", "C", "B", "B")
```

```python
def rush03(x, y):
    return draw_rush(x, y, "A", "C", "A", "C", "B", "B")
```

```python
def rush04(x, y):
    return draw_rush(x, y, "A", "C", "C", "A", "B", "B")
```

```python
def test_all(x, y):
    print("Rush00:")
    print(rush00(x, y))
    print("\nRush01:")
    print(rush01(x, y))
    print("\nRush02:")
    print(rush02(x, y))
    print("\nRush03:")
    print(rush03(x, y))
    print("\nRush04:")
    print(rush04(x, y))
```

```python
test_all(5, 3)
```

```python
print(rush00(1, 5))
print(rush01(1, 1))
print(rush02(4, 4))
```
