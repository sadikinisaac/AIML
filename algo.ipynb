```python
#Number of ways to climb stairs:
#If a person can climb either 1 stair or 2 stairs at a time, 
#for a given n stairs, 
#how many ways a person can reach the top from bottom?

def climbStairs(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    elif n == 2:
        return 2

    # Initialize an array to store the number of ways at each step
    ways = [0] * (n + 1)
    ways[1] = 1
    ways[2] = 2

    # Calculate the number of ways for each step
    for i in range(3, n + 1):
        ways[i] = ways[i - 1] + ways[i - 2]

    return ways[n]

# Example usage:
n = 4
result = climbStairs(n)
print(f"The number of ways to climb {n} stairs is: {result}")
```

```python
# What if a person can climb any number from a set X, 
# for example if X = {1,2,4}, 
# then he/she can climb 1, 2 or 4 stairs at a time. 
# How many ways he/she can climb from bottom to top?




def climbStairsWithSet(n, X):
    if n == 0:
        return 0

    # Initialize an array to store the number of ways at each step
    ways = [0] * (n + 1)
    ways[0] = 1  # There is 1 way to climb 0 stairs (doing nothing)

    # Calculate the number of ways for each step
    for i in range(1, n + 1):
        for x in X:
            if i - x >= 0:
                ways[i] += ways[i - x]

    return ways[n]

# Example usage with X = {1, 2, 4}
n = 5
X = {1, 2, 4}
result = climbStairsWithSet(n, X)
print(f"The number of ways to climb {n} stairs with set X = {X} is: {result}")
```
