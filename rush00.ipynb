```python
def ft_putchar(c):
    print(c, end="")
```

```python
def draw_rectangle(x, y, tl, tr, bl, br, h, v):
    if x <= 0 or y <= 0:
        return

    for row in range(1, y + 1):
        for col in range(1, x + 1):
            if row == 1 and col == 1:
                ft_putchar(tl)
            elif row == 1 and col == x:
                ft_putchar(tr)
            elif row == y and col == 1:
                ft_putchar(bl)
            elif row == y and col == x:
                ft_putchar(br)
            elif row == 1 or row == y:
                ft_putchar(h)
            elif col == 1 or col == x:
                ft_putchar(v)
            else:
                ft_putchar(" ")
        ft_putchar("\n")
```

```python
def rush00(x, y):
    draw_rectangle(x, y, 'o', 'o', 'o', 'o', '-', '|')
```

```python
def rush01(x, y):
    draw_rectangle(x, y, '/', '\\', '\\', '/', '*', '*')
```

```python
def rush02(x, y):
    draw_rectangle(x, y, 'A', 'A', 'C', 'C', 'B', 'B')
```

```python
def rush03(x, y):
    draw_rectangle(x, y, 'A', 'C', 'A', 'C', 'B', 'B')
```

```python
def rush04(x, y):
    draw_rectangle(x, y, 'A', 'C', 'C', 'A', 'B', 'B')
```

```python
print("Rush00 (5,3):")
rush00(5, 3)

print("\nRush01 (5,3):")
rush01(5, 3)

print("\nRush02 (5,3):")
rush02(5, 3)

print("\nRush03 (5,3):")
rush03(5, 3)

print("\nRush04 (5,3):")
rush04(5, 3)
```

```python
print("rush00(1,1):")
rush00(1,1)

print("\nrush02(1,5):")
rush02(1,5)

print("\nrush04(4,4):")
rush04(4,4)
```
